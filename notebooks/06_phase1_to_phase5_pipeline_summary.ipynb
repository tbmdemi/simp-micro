# 📘 Notebook 6: End-to-End Phase 1→5 Summary

This notebook joins the outputs from Phases 1 through 5 into a concise end-to-end pipeline report.

```python
import json
import sys
from pathlib import Path

import pandas as pd

sys.path.insert(0, str(Path.cwd() / "notebooks"))
import utils

root = Path.cwd()

# Load Phase 1 summaries
phase1_path = root / "outputs" / "pipeline" / "phase1" / "_all_summaries_parallel.json"
phase2_path = root / "outputs" / "multi_batch" / "decision_log.json"
phase3_path = root / "outputs" / "phase3" / "manifest.csv"
phase4_path = root / "outputs" / "phase4" / "evaluation_report.json"
phase5_path = root / "outputs" / "phase5" / "evaluation_report.json"

print('Phase 1 summary exists:', phase1_path.exists())
print('Phase 2 decision log exists:', phase2_path.exists())
print('Phase 3 manifest exists:', phase3_path.exists())
print('Phase 4 report exists:', phase4_path.exists())
print('Phase 5 report exists:', phase5_path.exists())
```

```python
# Load available summaries
if phase1_path.exists():
    with open(phase1_path, 'r') as f:
        phase1 = json.load(f)
    phase1_df = pd.DataFrame(phase1)
    print('Phase 1 summary rows:', len(phase1_df))
    display(phase1_df[['seed','n_samples','n_success','n_converged','best_obj_value']].head())

if phase2_path.exists():
    with open(phase2_path, 'r') as f:
        phase2 = json.load(f)
    print('\nPhase 2 entries:', len(phase2) if isinstance(phase2, list) else 'n/a')

if phase3_path.exists():
    phase3_df = pd.read_csv(phase3_path)
    print('\nPhase 3 manifest columns:', phase3_df.columns.tolist())
    print('Phase 3 rows:', len(phase3_df))

if phase4_path.exists():
    with open(phase4_path, 'r') as f:
        phase4 = json.load(f)
    print('\nPhase 4 overall metrics:', phase4.get('overall', {}).keys())

if phase5_path.exists():
    with open(phase5_path, 'r') as f:
        phase5 = json.load(f)
    print('\nPhase 5 property accuracy keys:', phase5.get('property_accuracy', {}).keys())
```

```python
# Build a concise pipeline summary table
summary = []
summary.append({'phase': 'Phase 1', 'artifact': 'LHS screening summaries', 'status': 'available'})
summary.append({'phase': 'Phase 2', 'artifact': 'Adaptive DOE decision log', 'status': 'available'})
summary.append({'phase': 'Phase 3', 'artifact': 'Manifest + NPZ datasets', 'status': 'available'})
summary.append({'phase': 'Phase 4', 'artifact': 'Surrogate evaluation report', 'status': 'available'})
summary.append({'phase': 'Phase 5', 'artifact': 'cVAE evaluation report', 'status': 'available'})
summary_df = pd.DataFrame(summary)
display(summary_df)
```
